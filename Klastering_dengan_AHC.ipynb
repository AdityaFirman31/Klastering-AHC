```python
import pandas as pd
from scipy.spatial.distance import pdist, squareform
from scipy.cluster.hierarchy import linkage, dendrogram
import matplotlib.pyplot as plt
```

```python
file_name = "Data AHC.xlsx"
data = pd.read_excel(file_name)
```

```python
data = data.iloc[1:, 1:].astype(float)
```

```python
manhattan_distance = pdist(data, metric='cityblock')
```

```python
complete_linkage = linkage(manhattan_distance, method='complete')
```

```python
average_linkage = linkage(manhattan_distance, method='average')
```

```python
plt.figure(figsize=(10, 5))
dendrogram(complete_linkage, labels=data.index + 1)
plt.title('Dendrogram (Complete Linkage)')
plt.xlabel('Siswa')
plt.ylabel('Jarak')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
dendrogram(average_linkage, labels=data.index + 1)
plt.title('Dendrogram (Average Linkage)')
plt.xlabel('Siswa')
plt.ylabel('Jarak')
plt.show()
```
